```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Resizing

# CNNTradFPool3
from utils.custom_layers import LowRankDense

model_CNNTradFPool3 = tf.keras.Sequential([
    Input(shape=(32, 32, 1)),
    Conv2D(filters=64, kernel_size=(20, 8), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(1, 3)),
    Conv2D(filters=64, kernel_size=(10, 4), activation='relu', padding='same'),
    MaxPooling2D(pool_size=(1, 1)),
    Flatten(),
    LowRankDense(units=32, rank=3),
    Dense(units=128, activation='relu'),
    Dense(units=10, activation='softmax')
])
```

```python
x = tf.random.normal((1, 32, 32, 1))

model_CNNTradFPool3.summary()

# plot model
tf.keras.utils.plot_model(model_CNNTradFPool3, to_file='CNNTradFPool3.png', show_shapes=True, show_layer_names=True)
```
