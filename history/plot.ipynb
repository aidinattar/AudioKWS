```python
import pandas as pd
```

```python
d = pd.read_pickle('CNNTradFPool3.pkl', )
loss = d['loss']
val_loss = d['val_loss']
acc = d['accuracy']
val_acc = d['val_accuracy']
epochs = range(len(loss))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid", palette="colorblind", font_scale=1.5)


fig, ax = plt.subplots(2,1,figsize=(5, 10))
markersize = 7
sns.lineplot(x=epochs, y=acc, ax=ax[0], label='Accuracy', marker='o', markersize=markersize, markeredgewidth=0)
sns.lineplot(x=epochs, y=val_acc, ax=ax[0], label='Validation Accuracy', marker='o', markersize=markersize, markeredgewidth=0)
sns.lineplot(x=epochs, y=loss, ax=ax[1], label='Loss', marker='o', markersize=markersize, markeredgewidth=0)
sns.lineplot(x=epochs, y=val_loss, ax=ax[1], label='Validation Loss', marker='o', markersize=markersize, markeredgewidth=0)


# fill area between lines
ax[0].fill_between(epochs, acc, val_acc, alpha=0.3, color='black')
ax[1].fill_between(epochs, loss, val_loss, alpha=0.3, color='black')
# remove space between subplots
fig.subplots_adjust(hspace=0.01)
ax[0].set_xlabel('')
ax[0].set_ylabel('Accuracy')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Loss')

fig.savefig('CNNTradFPool3.pdf', dpi=300, bbox_inches='tight')
```
